```python
movie_properties=ratings.groupby('movie_id').agg({'rating':[np.size,np.mean]})
```

```python
movie_dict={}
with open ('ml-100k/u.item',encoding='iso-8859-1') as f:
    for line in f:
        fields=line.rstrip('\n').split('|')
        movie_id=int(fields[0])
        name=fields[1]
        genres=fields[5:24]
        genres=list(map(int,genres))
        movie_dict[movie_id]=(name,genres,movie_properties.loc[movie_id].get(('rating','size')), 
                              movie_properties.loc[movie_id].get(('rating','mean')))
```

```python
#movie_dict[movie_id]=(name,genres,movie_properties.loc[movie_id].rating.get('size'), 
                              movie_properties.loc[movie_id].rating.get('mean'))
```

```python
movie_dict
```

```python
from scipy import spatial
def calculate_distance(a,b):
    genre_distance=spatial.distance.cosine(a[1],b[1])
    popularity_distance=abs(a[2]-b[2])
    return genre_distance+popularity_distance
```

```python
movie_dict[2]
```

```python
calculate_distance(movie_dict[2],movie_dict[4])
```

```python
dic_distances={}
for movie1 in movie_dict:
    list_distances=[]
    for movie2 in movie_dict:
        list_distances.append(calculate_distance(movie_dict[movie1],movie_dict[movie2]))
    dic_distances[movie1]=list_distances
```

```python
dic_distances
```

```python
df_distances=pd.DataFrame(dic_distances,columns=list(movie_dict.keys()))
df_distances.setindex=movie_dict[0]
```
