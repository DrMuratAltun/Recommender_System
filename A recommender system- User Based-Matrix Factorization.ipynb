```python
r_cols=['movie_id','title']
movies=pd.read_csv('ml-100k/u.item',encoding='iso-8859-1',sep='|',header=None,usecols=range(2),names=r_cols)
```

```python
movies.head()
```

```python
user=pd.read_table('ml-100k/u.user',sep='|',header=None,names=['Userid','age','gender','proffession','Zipcode'])
user.head()
```

```python
ratings=pd.merge(movies,ratings)
ratings.title.value_counts()
```

```python
ratings.movie_id.nunique()
```

```python
movie_ratings=ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
```

```python
movie_ratings
```

```python
starwars_rating=movie_ratings['Star Wars (1977)']
```

```python
starwars_rating.value_counts()
```

```python
movie_ratings[['101 Dalmatians (1996)', 'Star Wars (1977)']].corr()
```

```python
similar_movies=movie_ratings.corrwith(starwars_rating).sort_values(ascending=False)
```

```python
similar_movies
```

```python
similar_movies.dropna(inplace=True)
```

```python
similar_movies
```

```python
movie_stats=ratings.groupby('title').agg({'rating':[np.size,np.mean]})
movie_stats.head()
```

```python
popular_movies=movie_stats[movie_stats['rating']['size']>100]
```

```python
popular_movies.sort_values(('rating','mean'), ascending=False)
```

```python
popular_movies.columns
```

```python
type(similar_movies)
```

```python
similar_movies_df=pd.DataFrame(similar_movies,columns=['similarity'])
```

```python
popular_movies.columns=popular_movies.columns.get_level_values(0)
df=popular_movies.join(similar_movies_df)
```
